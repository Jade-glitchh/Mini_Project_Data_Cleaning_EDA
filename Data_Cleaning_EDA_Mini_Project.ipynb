# Superstore Sales Dataset - retail sales & profit analysis

#Import data analysis tools

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#Load Superstore Sales Dataset

```python
import pandas as pd
df = pd.read_csv("Superstore.csv", encoding='latin1')
df.head()
```

#Data cleaning - handling missing values

```python
df.dropna()  
```

```python
df.isna().sum()
```

#Data cleaning - standardizing categorical/text columns

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df
```

#Data cleaning - converting data types

```python
df['order_date'] = df['order_date'].astype(str).str.replace(r'[-.\s]', '/', regex=True)
df['ship_date'] = df['ship_date'].astype(str).str.replace(r'[-.\s]', '/', regex=True)

df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce', dayfirst=True)
df['ship_date'] = pd.to_datetime(df['ship_date'], errors='coerce', dayfirst=True)
```

 #Transform - create new columns 

```python
df['profit_margin'] = df['profit']/df['sales']
df.head()
```

#Transform - fix inconsistent  categories or mislabeled values

```python
columns = ['category', 'sub-category', 'region', 'customer_name']
df[columns] = df[columns].apply(lambda col: col.astype(str).str.strip().str.lower().str.title())
df
```

#Extract time-based features

```python
df['order_date'].dt.year
```

```python
df['order_date'].dt.strftime('%B')
```

```python
df['order_date'].dt.strftime('%A, %d')
```

#Data exploration

```python
df.describe()
```

#Data Visualization

```python
df['category'].value_counts()
```

```python
ax= df['category'].value_counts().plot(kind='bar', title='Category Distributions')
ax.set_xlabel('category')
ax.set_ylabel('Count')
plt.show
```

```python
df.plot(kind='scatter', x= 'sales', y= 'profit', title= 'Sales Vs. Profit')
plt.show()
```

```python
#using seaborn to group by
sns.scatterplot(x='sales', y='profit',hue='category', data=df)
```

```python
df['delivery_days'] = (df['ship_date'] - df['order_date']).dt.days

# Basic histogram of delivery times
sns.histplot(df['delivery_days'], bins=20, kde=True)
plt.title('Delivery Time Distribution')
plt.xlabel('Days to Ship')
plt.ylabel('Number of Orders')
plt.show()
```

```python
top_customers = df.groupby('customer_name').agg({'sales':'sum', 'profit':'sum'}).nlargest(10, 'sales')
top_customers.plot(kind='bar', figsize=(12, 6))
plt.title('Top 10 customer: Sale and Profit')
plt.show()
```
